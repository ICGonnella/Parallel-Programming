```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
time_comm_CBLAS = pd.read_csv('time_comm_CBLAS.dat', delimiter='\t',names=["rank", "size","dim","time"])
time_comm_GPU = pd.read_csv('time_comm_GPU.dat', delimiter='\t',names=["rank", "size","dim","time"])
time_comp_CBLAS = pd.read_csv('time_comp_CBLAS.dat', delimiter='\t',names=["rank", "size","dim","time"])
time_comp_GPU = pd.read_csv('time_comp_GPU.dat', delimiter='\t',names=["rank", "size","dim","time"])
```

```python
time_comm_CBLAS_mean = [np.mean(
    [time_comm_CBLAS["time"][i] for i in range(len(time_comm_CBLAS)) if time_comm_CBLAS["size"][i]==j]) 
                        for j in [1,2,5,8,10]]
time_comp_CBLAS_mean = [np.mean(
    [time_comp_CBLAS["time"][i] for i in range(len(time_comp_CBLAS)) if time_comp_CBLAS["size"][i]==j]) 
                        for j in [1,2,5,8,10]]
time_comm_GPU_mean = [np.mean(
    [time_comm_GPU["time"][i] for i in range(len(time_comm_GPU)) if time_comm_GPU["size"][i]==j]) 
                        for j in [1,2,5,8,10]]
time_comp_GPU_mean = [np.mean(
    [time_comp_GPU["time"][i] for i in range(len(time_comp_GPU)) if time_comp_GPU["size"][i]==j]) 
                        for j in [1,2,5,8,10]]
```

```python
plt.bar([1,2,5,8,10],[time_comp_CBLAS_mean[i]+time_comm_CBLAS_mean[i] for i in range(5)], color="green", label="Communication Time")
plt.bar([1,2,5,8,10],time_comp_CBLAS_mean, color="red",label="Computational Time")
plt.grid()
plt.xticks([1,2,5,8,10])
plt.legend()
plt.show()
```

```python
plt.bar([1,2,5,8,10],[time_comp_GPU_mean[i]+time_comm_GPU_mean[i] for i in range(5)], color="green", label="Communication Time")
plt.bar([1,2,5,8,10],time_comp_GPU_mean, color="red",label="Computational Time")
plt.grid()
plt.xticks([1,2,5,8,10])
plt.legend()
plt.show()
```
