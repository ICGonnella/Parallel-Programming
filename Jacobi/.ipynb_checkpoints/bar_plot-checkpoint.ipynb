```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
time_comm = pd.read_csv('make_results/time_comm.dat', delimiter='\t',names=["rank", "size","dim", "iterations","time"])
time_comm_acc = pd.read_csv('makeacc_results/time_comm.dat', delimiter='\t',names=["rank", "size","dim", "iterations","time"])
time_comp = pd.read_csv('make_results/time_comp.dat', delimiter='\t',names=["rank", "size","dim", "iteration","time"])
time_comp_acc = pd.read_csv('makeacc_results/time_comp.dat', delimiter='\t',names=["rank", "size","dim", "iterations","time"])
```

```python
time_comm
```

```python
time_comm_mean = [np.min(
    [time_comm["time"][i] for i in range(len(time_comm)) 
     if np.logical_and(time_comm["size"][i]==j, time_comm["dim"][i]==10000)])
                        for j in [1,2,5,8,10]]
time_comm_acc_mean = [np.min(
    [time_comm_acc["time"][i] for i in range(len(time_comm_acc)) 
     if np.logical_and(time_comm_acc["size"][i]==j, time_comm_acc["dim"][i]==10000)])
                        for j in [1,2,5,8,10]]
time_comp_mean = [np.min(
    [time_comp["time"][i] for i in range(len(time_comp)) 
     if np.logical_and(time_comp["size"][i]==j, time_comp["dim"][i]==10000)])
                        for j in [1,2,5,8,10]]
time_comp_acc_mean = [np.min(
    [time_comp_acc["time"][i] for i in range(len(time_comp_acc)) 
     if np.logical_and(time_comp_acc["size"][i]==j, time_comp_acc["dim"][i]==10000)])
                        for j in [1,2,5,8,10]]
```

```python
width=0.3
plt.bar([1+width/2,2+width/2,5+width/2,8+width/2,10+width/2],[time_comm_mean[i] for i in range(5)], color="green", label="Communication Time", width=width)
plt.bar([1-width/2,2-width/2,5-width/2,8-width/2,10-width/2],time_comp_mean, color="red",label="Computational Time", width=width)
plt.grid()
plt.xticks([1,2,5,8,10])
plt.legend()
plt.show()
```

```python
width=0.3
plt.bar([1+width/2,2+width/2,5+width/2,8+width/2,10+width/2],[time_comm_acc_mean[i] for i in range(5)], color="green", label="Communication Time", width=width)
plt.bar([1-width/2,2-width/2,5-width/2,8-width/2,10-width/2],time_comp_acc_mean, color="red",label="Computational Time", width=width)
plt.grid()
plt.xticks([1,2,5,8,10])
plt.legend()
plt.show()
```
